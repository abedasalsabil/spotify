```python
# Import the modules
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Read in the CSV file and create the Pandas DataFrame
df_spotify_April = pd.read_csv(
    Path("Resources/April_tracks_metadata.csv")
)

# Review the DataFrame
df_spotify_April.head()
```

```python
df = df_spotify_April[['track.album.name','track.name','track.album.album_type',
                       'track.album.artists','track.popularity','danceability',
                      'energy', 'key', 'loudness','speechiness', 'acousticness', 
                      'instrumentalness', 'liveness', 'valence', 'tempo']].copy()
df
```

```python
df['artist'] = df['track.album.artists'].str.extract(r"'name': '(.+?)'")
df.drop('track.album.artists', axis=1, inplace=True)
df
```

```python
artist_popularity = df.groupby('artist')['track.popularity'].sum().reset_index()
```

```python
top_artists = artist_popularity.nlargest(20, 'track.popularity')
```

```python
import matplotlib.pyplot as plt

plt.bar(top_artists['artist'], top_artists['track.popularity'])
plt.xlabel('Artist')
plt.ylabel('track.popularity')
plt.title('Top 10 Popular Artists of April')
plt.xticks(rotation=90)
plt.show()
```
